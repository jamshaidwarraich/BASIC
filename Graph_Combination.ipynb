```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)

# Calculate y values for the function x
y_values_x = x_values

# Plot the function x
plt.figure(figsize=(6, 4))
plt.plot(x_values, y_values_x, label='x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate y values for the function 2x
y_values_2x = 2 * x_values

# Plot the function 2x
plt.figure(figsize=(6, 4))
plt.plot(x_values, y_values_2x, label='2x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of 2x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate y values for the function sin(x)
y_values_sin = np.sin(x_values)

# Plot the function sin(x)
plt.figure(figsize=(6, 4))
plt.plot(x_values, y_values_sin, label='sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a new figure and axis for the combined plots
plt.figure(figsize=(15, 5))

# Plot x in the first subplot
plt.subplot(1, 3, 1)
plt.plot(x_values, y_values_x, label='x')
plt.title('x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Plot 2x in the second subplot
plt.subplot(1, 3, 2)
plt.plot(x_values, y_values_2x, label='2x')
plt.title('2x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Plot sin(x) in the third subplot
plt.subplot(1, 3, 3)
plt.plot(x_values, y_values_sin, label='sin(x)')
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)

# Calculate y values for the function x
y_values_x = x_values

# Calculate y values for the function 2x
y_values_2x = 2 * x_values

# Calculate y values for the function sin(x)
y_values_sin = np.sin(x_values)

# Plot all functions on a single figure
plt.figure(figsize=(12, 6))

# Plot x
plt.plot(x_values, y_values_x, label='x')

# Plot 2x
plt.plot(x_values, y_values_2x, label='2x')

# Plot sin(x)
plt.plot(x_values, y_values_sin, label='sin(x)')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x, 2x, and sin(x)')
plt.legend()  # Show legend with function names

plt.grid(True)  # Add grid

plt.show()
```
